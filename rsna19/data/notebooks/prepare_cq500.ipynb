```python
paths = []
uids = []
for path in iglob('/kolos/m2/ct/data/CQ500/*/*/dicom/'):
    for dicom_path in Path(path).iterdir():
        dicom_metadata = pydicom.dcmread(str(dicom_path), stop_before_pixels=True)
        uids.append(dicom_metadata.SOPInstanceUID)
        paths.append(str(dicom_path))
```

```python
anys = []
epidural = []
intraparenchymal = []
intraventricular = []
subarachnoid = []
subdural = []

new_paths = []
path_ends = []
study_ids = []
old_paths = []
for path in iglob('/kolos/m2/ct/data/CQ500/*/*/dicom/*'):
    dicom_boxes = df.loc[df['path'] == path]
    bleedings = [bleeding for bleeding in dicom_boxes.labelName.unique() if bleeding != 'Chronic']
    if len(bleedings) > 0:
        anys.append(1)
        
        if 'Subarachnoid' in bleedings:
            subarachnoid.append(1)
        else:
            subarachnoid.append(0)
            
        if 'Epidural' in bleedings:
            epidural.append(1)
        else:
            epidural.append(0)
            
        if 'Intraparenchymal' in bleedings:
            intraparenchymal.append(1)
        else:
            intraparenchymal.append(0)
            
        if 'Intraventricular' in bleedings:
            intraventricular.append(1)
        else:
            intraventricular.append(0)
            
        if 'Subdural' in bleedings:
            subdural.append(1)
        else:
            subdural.append(0)
    else:
        anys.append(0)
        epidural.append(0)
        intraparenchymal.append(0)
        intraventricular.append(0)
        subarachnoid.append(0)
        subdural.append(0)
    
    path = Path(path)
    scan_name = f"{str(path.parts[6]).split('_')[0]}_{str(path.parts[7])}"
    slice_num = int(str(path.name).split('_')[-1][2:].split('.')[0])
    path_end = Path(scan_name) / 'dicom' / f'{slice_num:03d}.dcm'
    out_path = data_root / path_end

    path_ends.append(path_end)
    new_paths.append(out_path)
    old_paths.append(str(path))
    
    if len(dicom_boxes) > 0:
        study_ids.append(dicom_boxes.StudyInstanceUID.values[0])
    else:
        dicom_metadata = pydicom.dcmread(str(path), stop_before_pixels=True)
        study_ids.append(dicom_metadata.StudyInstanceUID)
        
df_cq = pd.DataFrame.from_dict({'any': anys,
                               'epidural': epidural,
                               'intraparenchymal': intraparenchymal,
                               'intraventricular': intraventricular,
                               'subarachnoid': subarachnoid,
                               'subdural': subdural,
                               'new_path': new_paths,
                               'relative_path': path_ends,
                               'path': old_paths,
                               'study_id': study_ids})

df_cq.head()
```

```python
grouped_by_patient = df_cq.groupby('study_id', group_keys=True)
```

```python
import random
NUM_FOLDS = 5

curr_foldnum = 0
ids = []
folds = []

patient_ids = list(grouped_by_patient.groups.keys())
random.shuffle(patient_ids)
for idx, patient_id in enumerate(patient_ids):
    patient_slices = grouped_by_patient.groups[patient_id]
                                 
    if (idx+1) % (len(grouped_by_patient.groups)//NUM_FOLDS) == 0 and curr_foldnum < (NUM_FOLDS-1):
        curr_foldnum += 1
    for slice_id in patient_slices:
        ids.append(slice_id)
        folds.append(curr_foldnum)
```

```python
cum_count = 0
for num in range(NUM_FOLDS):
    count =0 
    for val in folds:
        if val == num:
            count +=1
    cum_count += count
    print(num, count)

print(cum_count == len(df_cq))
```

```python
folds = pd.DataFrame.from_dict({'id': ids, 'fold': folds})
folds = folds.set_index('id')
df_cq = df_cq.join(folds)
df_cq.head(20)
```

```python
df_cq.to_csv('cq500_paths_labels_folds.csv', index=False)
```
